```python
def resonanceterm(gamma,delta):
    return np.divide(np.multiply(511,gamma),np.add(np.power(delta,2),np.divide(np.power(gamma,2),4)))

def t0vbb(NME, PSF, meff=1, gamma=0, delta=0):
    if gamma>0 and delta>0:
        #calculate 0vecec half-life
        return 0
    else:
        #calculate 0vecb+ or 0vb+b+ half-life
        return 1
```

```python
x=np.arange(0,15,0.01)
delta=1.86 #keV
deltaerr=0.15 #keV
gamma=0.0198 #keV

plt.plot(x,resonanceterm(gamma,x), color='crimson', label='Resonance curve')
plt.yscale('log')
plt.axvline(delta, linestyle='-', color='#29BF12', label='Nesterenko et al. (2012)')
plt.axvspan(delta-deltaerr,delta+deltaerr, facecolor='#29BF12', edgecolor='none', alpha=0.6)
plt.axvspan(delta-2*deltaerr,delta+2*deltaerr, facecolor='#29BF12', edgecolor='none', alpha=0.4)

plt.xlabel('$\\Delta$ [keV]')
plt.ylabel('$\\frac{m_{\\rm{e}}c^2\\Gamma}{\\Delta^2+\Gamma^2/4}$')
plt.xlim(0,3)
plt.ylim(1e0,2e5)
plt.legend(ncol=1, frameon=False, loc='upper left')
plt.draw()
plt.savefig('0vecec_resonance.pdf', Transparent=True, bbox_inches='tight')

plt.show()
```
